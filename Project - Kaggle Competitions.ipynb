```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
import time
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore')
```

```python
def getKaggleCompetitions(page_url, driver):
    
    data = {'title': [], 'price': [], 'launch': [], 'close': [], 'tags': [], 'teams': [], 'competitors': [], 'entries': [], 'description': [], 'files': [], 'size': [], 'points': [], 'tiers': [], 'difficulty': []}
    competitions = pd.DataFrame(data)
    driver.get(page_url)
    
    #click on sign in
    sign_in = driver.find_element(By.XPATH, "//main[1]/div[1]/div[1]/div[4]/div[2]/div[2]/div[1]/div[1]/a[1]/button[1]")
    sign_in.click()
    time.sleep(3)
    
    #click on sign in with email from options
    email = driver.find_element(by=By.XPATH, value='//a[@href="/account/login?phase=emailSignIn&returnUrl=%2Fcompetitions"]')
    email.click()
    time.sleep(3)
    
    #fill in user name and password
    username = driver.find_element(By.NAME, "email")
    password = driver.find_element(By.NAME, "password")
    username.send_keys("preetireddykoppulakc@gmail.com")
    password.send_keys("ramanthapur")
    time.sleep(3)

    #submit user name and password
    submit = driver.find_element(By.XPATH, "//main[1]/div[1]/div[1]/div[3]/form[1]/div[2]/div[3]/button[1]")
    submit.click()
    
    time.sleep(3)
            
    load_more_1=driver.find_element(By.XPATH, "//main[1]/div[1]/div[1]/div[6]/div[2]/div[4]/div[1]/div[2]/div[1]/div[1]/button[1]")
    load_more_1.click()
    time.sleep(2)
    
    for j in range(28):
        if j==1:
            load_more=driver.find_element(By.XPATH, "//main[1]/div[1]/div[1]/div[6]/div[2]/div[4]/div[1]/div[2]/div[1]/div[1]/button[1]")
            load_more.click()
            time.sleep(2)
            
        for i in range(j):
            list_path = driver.find_element(By.XPATH, '//main[1]/div[1]/div[1]/div[6]/div[2]/div[5]/div[1]/div[1]/div[1]/div[2]/i[2]')
            list_path.click()
            time.sleep(2)

        for i in range(20):
            #open all competitions
            if j==0 and i>0:
                load_more_3=driver.find_element(By.XPATH, "//main[1]/div[1]/div[1]/div[6]/div[2]/div[4]/div[1]/div[2]/div[1]/div[1]/button[1]")
                load_more_3.click()
                time.sleep(2)

            list_path = '//ul[1]/li[{0}]/div[1]/a[1]/div[2]/div[1]'.format(i+1)
            comp = driver.find_element(By.XPATH, list_path)
            comp.click()
            time.sleep(5)
            try:
                d = {'title': '', 'price': '', 'files': '', 'size': '', 'launch': '', 'teams': '', 'competitors': '', 'entries': '', 'description': '', 'tags': '', 'points': '', 'tiers': '', 'difficulty': ''}
                try:
                    title = driver.find_element(By.CSS_SELECTOR, 'h1.competition-header__title')
                    d['title'] = title.text
                except: 
                    time.sleep(2)
                try:
                    price = driver.find_element(By.CSS_SELECTOR, 'span.competition-header__prize-value')
                    d['price'] = price.text
                except:
                    time.sleep(2)
                try:
                    difficulty = driver.find_element(By.CSS_SELECTOR, 'p.competition-header__classification-text')
                    d['difficulty'] = difficulty.text
                except:
                    time.sleep(2)
                try:
                    points = driver.find_element(By.XPATH, '//main[1]/div[1]/div[1]/div[6]/div[2]/div[1]/div[2]/div[3]/div[2]/div[1]/p[2]')
                    d['points'] = points.text
                except:
                    time.sleep(2)
                try:
                    tiers = driver.find_element(By.XPATH, '//main[1]/div[1]/div[1]/div[6]/div[2]/div[1]/div[2]/div[3]/div[2]/div[2]/p[2]')
                    d['tiers'] = tiers.text
                except:
                    time.sleep(2)
                try:
                    launch = driver.find_element(By.XPATH, "//main[1]/div[1]/div[1]/div[6]/div[2]/div[1]/div[2]/div[2]/div[1]/div[1]/div[1]/div[1]/div[1]/div[3]/div[1]/span[2]")
                    d['launch'] = launch.text
                    close = driver.find_element(By.XPATH, "//main[1]/div[1]/div[1]/div[6]/div[2]/div[1]/div[2]/div[2]/div[1]/div[1]/div[1]/div[1]/div[1]/div[4]/div[1]/span[2]")
                    d['close'] = close.text
                    time.sleep(2)
                except:
                    time.sleep(2)
                try:
                    teams = driver.find_element(By.XPATH, "//main[1]/div[1]/div[1]/div[6]/div[2]/div[1]/div[2]/div[3]/div[1]/div[1]/p[1]/span[1]")
                    d['teams'] = teams.text
                    competitors = driver.find_element(By.XPATH, "//main[1]/div[1]/div[1]/div[6]/div[2]/div[1]/div[2]/div[3]/div[1]/div[2]/p[1]/span[1]")
                    d['competitors'] = competitors.text
                    entries = driver.find_element(By.XPATH, "//main[1]/div[1]/div[1]/div[6]/div[2]/div[1]/div[2]/div[3]/div[1]/div[3]/p[1]/span[1]")
                    d['entries'] = entries.text
                    time.sleep(2)
                except:
                    time.sleep(2)
                try:
                    description = driver.find_element(By.XPATH, "//main[1]/div[1]/div[1]/div[6]/div[2]/div[1]/div[2]/div[1]/div[2]/div[1]/div[2]/div[1]/div[1]")
                    d['description'] = description.text
                    time.sleep(2)
                except:
                    time.sleep(2)
                try:
                    tags = driver.find_element(By.XPATH, "//main[1]/div[1]/div[1]/div[6]/div[2]/div[1]/div[3]/div[1]/div[1]/div[1]/div[2]")
                    d['tags'] = tags.text.split('\n')
                    time.sleep(2)
                except:
                    time.sleep(2)
                try:
                    data_tab = driver.find_element(By.XPATH, "//main[1]/div[1]/div[1]/div[6]/div[2]/div[1]/div[1]/div[1]/div[1]/div[2]/div[2]/div[1]/div[1]/div[1]/div[1]/nav[1]/div[1]/a[2]")
                    if data_tab.text == 'Data':
                        data_tab.click()
                        time.sleep(2)
                        files = driver.find_element(By.XPATH, "//main[1]/div[1]/div[1]/div[6]/div[2]/ div[1]/div[2]/div[2]/p[1]")
                        d['files'] = files.text
                        size = driver.find_element(By.XPATH, "//main[1]/div[1]/div[1]/div[6]/div[2]/ div[1]/div[2]/div[2]/p[2]")
                        d['size'] = size.text
                        driver.back()
                        time.sleep(2)
                except:
                    time.sleep(2)
                competitions = competitions.append(d, ignore_index = True)
                time.sleep(2)
                driver.back()
                time.sleep(2)
            except:
                driver.back()
                time.sleep(2)

    return competitions
```

```python
# Test the function

executable_path = 'C:/Users/preet/Downloads/driver/chromedriver_win32/chromedriver.exe'
driver = webdriver.Chrome(executable_path=executable_path)

page_url = 'https://www.kaggle.com/competitions'
competitions = getKaggleCompetitions(page_url, driver)
competitions
driver.quit()
```
